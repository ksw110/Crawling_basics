### requests
- 웹 문서를 요청하거나 응답받을 때 사용하는 라이브러리(모듈)
- url 주소를 입력하여 해당 페이지의 html 정보를 요청하고 응답받는다

```python
import requests as req
```

```python
url = 'https://www.naver.com/'

# url에 저장되어 있는 웹 문서 요청하기 : get()
# Response[200] : 성공적으로 요청/응답이 진행된 상태
res = req.get(url) # res 라는 변수 안에 불러온 req(requests)에서 get(가져오겠다) url을
res
```

```python
# 응답받은 데이터(res)에서 웹 문서 접근하기
res.text
```

### BeautifulSoup (bs)
- 웹 문서 안에 있는 데이터를 추출할 수 있도록 함수를 제공해주는 모듈
- 1. req를 통해서 html 정보를 받아옴 (데이터정보 - 텍스트) --> 파이썬객체로 변환
- 2. bs 제공하는 함수를 통해 원하는 태그의 정보에 접근하여 데이터를 추출할 수 있다

```python
from bs4 import BeautifulSoup as bs
```

```python
# req를 통해 html 정보(텍스트형태) --> 파이썬객체로 변환(태그 접근)
html = bs(res.text, 'lxml') # 파이썬으로 변환할 때 'lxml' 꼭 써줘야 함 (파싱 parsing)
html.text
```

```python
# css 선택자를 활용하여 접근할 수 있게 하는 함수 : .select_one()
# .select_one().text --> 태그 안에 있는 텍스트만 가져올 때 사용
# html.select_one(css 선택자): css선택자에 맞는 가장 먼저 찾아지는 첫번째 html 객체를 접근
# css선택자
# 태그 선택자 : 태그명
# 아이디 선택자 : #아이디명
# 클래스 선택자 : .클래스명
# 계층 선택자 : 자식(>), 자손(띄어쓰기), 형제(~), 인접형제(+)
html.select_one('title').text
```

```python
# html.select(css선택자): css선택자에 맞는 html 객체 모두 접근
# 결과가 리스트 형식으로 반환
html.select('title')[0].text
```

### 네이버 통합검색 페이지에서 데이터 수집하기
 - 검색키워드 : 겨울축제

```python
# 1. url2
url2 = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EA%B2%A8%EC%9A%B8%EC%B6%95%EC%A0%9C'
# 2. get() html 요청
res2 = req.get(url2)
# 3. bs 파이썬 객체 변환 -> 겨울축제 이름들 추출
html2 = bs(res2.text, 'lxml')
html2.select_one('title').text
```

```python
# 4. -> 겨울축제 이름들 추출
# 1개만 가져와볼래 ! => select_one()
html2.select_one('.this_text > a')
```

```python
# 4개 다 가져와보기!
for i in range(len(html2.select('.this_text > a'))):
    print(html2.select('.this_text > a')[i].text)
```

```python
html2_lst = html2.select('.this_text > a')
for i in html2_lst:
    print(i.text)
```

```python
# href 속성값 접근
link = html2_lst[0].get('href')
link2 = 'https://search.naver.com/search.naver' + link
print(link2)
```

### 네이버 뉴스 기사 크롤링하기
  - 뉴스 제목, 내용 수집한 후 출력해보기
  - 1. 어떤 태그에 있는지
  - 2. 어떤 클래스 등등 속성으로 이루어져 있는지

```python
news_url = 'https://n.news.naver.com/article/660/0000023947?cds=news_media_pc&type=editn'
```

```python
res = req.get(news_url)
res
# 에러 > 컴퓨터에게 브라우저에서 요청했다는 인식을 설정할 필요가 있음
```

```python
# 브라우저 요청 인식 설정
# 모든 페이지를 요청할 때 header값 지정할 것!
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
res = req.get(news_url, headers = header)
res
```

```python
# 파이썬 객체 변환
news_html = bs(res.text, 'lxml')
```

```python
news_html
```

```python
# 제목 내용 strip = \n 개행 지우는것
for i in range(len(news_html.select('#title_area, #dic_area'))):
    print(news_html.select('#title_area, #dic_area')[i].text.strip())
```

```python
# 뉴스 제목, 뉴스내용
title = news_html.select_one('#title_area > span')
content = news_html.select_one('#dic_area')
```

```python
print(title.text)
print(content.text.strip())
```
